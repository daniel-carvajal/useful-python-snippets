## define una **lista**...

```
cosas = ['carro', 'zapato', 'cuaderno', 'cuarentaYcinco']

print(cosas)
```

## recorre lista con **"for loop"**
El bucle **for** se utiliza para recorrer los elementos de un objeto iterable (lista, tupla, conjunto, diccionario, …)

```
for x in cosas:
  print(x)
```

```
cosas_copy = []
for x in cosas:
  cosas_copy.append(x)

print(cosas_copy)
```

## ahora con '**list comprehension**' embes de 'for loop'...

```
cosas_copy_v2 = [x for x in cosas]
print(cosas_copy_v2)
```

list comprehension pero modificando a X (hacemos cada input mayusucula)

```
cosas_copy_upper = [x.upper() for x in cosas]
print(cosas_copy_upper)
```

y ahora list comprehension con condiciones **"if"**

```
cosas_copy_upper_v2 = [x.upper() for x in cosas if len(x) > 5]
print(cosas_copy_upper_v2)
```

##mas exemplos...

```
numerosRandom = [123 , 1233 , 4123 , 4423, 5324, 9643]
numerosRandomCambio = [x**2 for x in numerosRandom]
print(numerosRandomCambio)
```
